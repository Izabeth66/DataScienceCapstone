```python
# TASK 3: Calculate the distances between a launch site to its proximities
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371
    return c * r

# Example launch site (first one in dataset)
launch_site = spacex_df.iloc[0]
launch_coords = [launch_site['Lat'], launch_site['Long']]

# Define proximities (replace with actual ones if needed)
rail_station = [28.562302, -80.570773]
highway = [28.563197, -80.576820]
coastline = [28.56342, -80.5678]
city = [28.6084, -80.6043]

# List of landmarks with names
landmarks = {
    'Rail Station': rail_station,
    'Highway': highway,
    'Coastline': coastline,
    'City': city
}

# Initialize map
distance_map = folium.Map(location=launch_coords, zoom_start=12)

# Add marker for launch site
folium.Marker(launch_coords, popup='Launch Site', icon=folium.Icon(color='blue')).add_to(distance_map)

# Add landmarks and draw lines
for name, coords in landmarks.items():
    # Add marker
    folium.Marker(coords, popup=name, icon=folium.Icon(color='gray')).add_to(distance_map)
    
    # Draw line
    folium.PolyLine([launch_coords, coords], color='orange', weight=2.5, dash_array='5').add_to(distance_map)
    
    # Show distance as popup
    distance_km = haversine(launch_coords[0], launch_coords[1], coords[0], coords[1])
    mid_point = [(launch_coords[0] + coords[0]) / 2, (launch_coords[1] + coords[1]) / 2]
    folium.Marker(mid_point, 
                  icon=folium.DivIcon(html=f"""<div style="font-size: 10pt">{distance_km:.2f} km</div>""")
                 ).add_to(distance_map)

# Show map
distance_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
